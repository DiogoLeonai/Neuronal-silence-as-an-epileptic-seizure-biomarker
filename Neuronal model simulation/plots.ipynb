```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
import csv
import time
```

In the cell below, we are importing the datas

```python
df_CV = pd.read_csv('CVt_gsyn0.001835.dat', names = ['time', 'CV'], sep='\t')
df_R = pd.read_csv('Rt_gsyn0.001835.dat', names = ['time', 'R'], sep='\t')
df_T = pd.read_csv('T_gsyn0.001835.dat', names = ['time', 'T'], sep='\t')
df_raster = pd.read_csv('Raster_gsyn0.001835.dat', names = ['time', 'neuron'], sep='\t')    
```

The CV, R and T are plotted in the cell below

```python
plt.figure(figsize=(20, 6))

###### CV and R #############
plt.subplot(2, 1, 1)
plt.plot(df_CV['time']/1000, df_CV['CV'], label='CV', color='blue')
plt.plot(df_R['time']/1000, df_R['R'], label='R', color='red')
plt.xlabel('Time (s)', fontsize = 20)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(0,250)

########## Silence duration <T> ########
plt.subplot(2, 1, 2)
plt.plot(df_T['time']/1000, df_T['T'], label='T', color='green')
plt.xlabel('Time (s)', fontsize = 20)
plt.ylabel('<T> (ms)', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(0,250)

plt.tight_layout()
```

The raster plot and the silence time of a transition from a down state to an up state is displayed below

```python
plt.figure(figsize=(14, 9))

time_interval = [53.0, 56.0] # Adjust this to your desired time interval

###### Raster plot ######
plt.subplot(2,1,1)
plt.scatter(df_raster['time']/1000, df_raster['neuron'], color='black', s=1)
plt.xlabel('Time (s)', fontsize = 20)
plt.ylabel('i', fontsize = 20)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(time_interval[0], time_interval[1])
plt.ylim(0,200)

###### T ################
plt.subplot(2, 1, 2)
plt.axhline(y=120, color='red', linestyle='--', c= 'black')
plt.plot(df_T['time']/1000, df_T['T'], label='T', color='green', lw= 5)

plt.xlabel('Time (s)', fontsize = 20)
plt.ylabel('<T> (ms)', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(time_interval[0], time_interval[1])
plt.tight_layout()
```
