```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
import csv
import time
```

```python
df_CV = pd.read_csv('CVt_gsyn0.001835.dat', names = ['time', 'CV'], sep='\t')
df_R = pd.read_csv('Rt_gsyn0.001835.dat', names = ['time', 'R'], sep='\t')
df_T = pd.read_csv('T_gsyn0.001835.dat', names = ['time', 'T'], sep='\t')
```

```python
plt.figure(figsize=(20, 6))

plt.subplot(2, 1, 1)
plt.plot(df_CV['time']/1000, df_CV['CV'], label='CV', color='blue')
plt.plot(df_R['time']/1000, df_R['R'], label='R', color='red')
plt.xlabel('Time (s)', fontsize = 20)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(0,250)

plt.subplot(2, 1, 2)
plt.plot(df_T['time']/1000, df_T['T'], label='T', color='green')
plt.xlabel('Time (s)', fontsize = 20)
plt.ylabel('<T> (ms)', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlim(0,250)

plt.tight_layout()
```
